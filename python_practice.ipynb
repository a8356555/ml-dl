```
# 通過字典進行分組
people = pd.DataFrame(np.random.randn(5, 5),
                    columns=['a', 'b', 'c', 'd', 'e'],
                    index=['Joe', 'Steve', 'Wes', 'Jim', 'Travis'])
mapping = {'a': 'red', 'b': 'red', 'c': 'blue',
           'd': 'blue', 'e': 'red', 'f' : 'orange'}
people.groupby(mapping, axis=1).mean()                


# 應用於分位桶
frame = pd.DataFrame({'data1': np.random.randn(1000),
                      'data2': np.random.randn(1000)})
quantiles = pd.cut(frame.data1, 4) #獲得分組array

def get_stats(group):
  return {'min': group.min(), 'max':group.max()}


grouped = frame.groupby(quantiles)['data2']

grouped.apply(get_stats).unstack()

#用分組平均填充空值
states = ['Ohio', 'New York', 'Vermont', 'Florida',
          'Oregon', 'Nevada', 'California', 'Idaho']

group_key = ['East'] * 4 + ['West'] * 4

data = pd.Series(np.random.randn(8), index=states)
data[['Vermont', 'Nevada', 'Idaho']] = np.nan
fill_mean = lambda g: g.fillna(g.mean())
data.groupby(group_key).apply(fill_mean)
```

```
from dateutil.parser import parse
parse('2011-01-03')
parse('Jan 31, 1997 10:45 PM')
parse('6/12/2011', dayfirst=True)

longer_ts = pd.Series(np.random.randn(1000),
                   index=pd.date_range('1/1/2000', periods=1000))
longer_ts['2001']
longer_ts['2001-01']
longer_ts['2001-01-05':'2001-01-20']

dates = pd.DatetimeIndex(['1/1/2000', '1/2/2000', '1/2/2000',
                          '1/2/2000', '1/3/2000'])
dup_ts = pd.Series(np.arange(5), index=dates)
grouped = dup_ts.groupby(level=0)
grouped.mean()

pd.date_range('2000-01-01', periods=10, freq='1h30min')


rng = pd.date_range('2000-01-01', periods=12, freq='T')
ts = pd.Series(np.arange(len(rng)), index=rng)
ts.resample('5min', closed = 'right').sum()

window = pd.date_range('2000-01-01', periods=300, freq='D')
ts = pd.Series(np.arange(len(window)), index=window)
ts.rolling(2).mean()
```

Ch12 pandas的高級應用
  1. category
    * 增加效能:在数据仓库中，最好的方法是使用所谓的包含不同值的维表(Dimension Table)，将主要的参数存储为引用维表整数键
  2. transform
    * 它可以产生向分组形状广播标量值
    * 它可以产生一个和输入组形状相同的对象
    * 它不能修改输入
    * vs apply: apply若使用聚合函式不會廣播，只會依照分組獲得結果
    * transform與apply獲得同樣結果:當運算為針對每個元素eg. lambda x:x-x.mean()
    * 解封:內置函式(mean, sum...) + 矢量化運算 會比把運算都丟在custom_func=> transform(custom_func) 快
  3. 鍊式編程技術 .groupby().apply()....
    * 只是一種習慣
    * 拆開可讀性比較高
    * 使用.pipe()更容易鍊式: func(df) => df.pipe(func1).pipe(func2) ...


```
df = pd.DataFrame({'key': ['a', 'b', 'c'] * 4,
                   'value': np.arange(12.)})

df.groupby('key')['value'].apply(lambda x:x.mean())
df.groupby('key')['value'].transform(lambda x:x.mean())

g = df.groupby('key')['value']
def one():
  g.transform(lambda x:(x-x.mean())/x.std())
def two():
  (df['value'] - g.transform('mean'))/g.transform('std')
%time one()
%time two()
```
